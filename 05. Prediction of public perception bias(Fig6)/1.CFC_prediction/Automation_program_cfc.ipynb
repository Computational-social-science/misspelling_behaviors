```python
import os
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

def merge_excel_sheets(input_folder, output_file):
    # 获取文件夹中的所有 Excel 文件
    excel_files = [file for file in os.listdir(input_folder) if file.endswith('.xlsx')]

    # 创建一个 Excel writer 对象来将所有工作表写入一个 Excel 文件
    with pd.ExcelWriter(output_file) as writer:
        # 遍历每个 Excel 文件
        for excel_file in excel_files:
            df = pd.read_excel("./data/"+excel_file, sheet_name="Sheet1")
            
            df['all'] = df.iloc[:, 1:].sum(axis=1)
            df = df.drop(df.columns[2:-1], axis=1)
            
            # 创建一个MinMaxScaler对象
            scaler = MinMaxScaler()

            # 假设df是你的DataFrame，列名为col2和col3
            cols_to_normalize = [excel_file.split(',')[0], 'all']
            
            
            #对起始点前的数据清洗清零
            word_number_pairs = {}
            with open('data_new.txt', 'r') as f:
                for line in f:
                    word = line.strip().split(',')[0]
                    number = line.strip().split(' ')[1]
                    word_number_pairs[word] = int(number)
                   
            if cols_to_normalize[0] in word_number_pairs:
                number = word_number_pairs[cols_to_normalize[0]]
                print(number)
                # 获取需要清零的索引
                df.iloc[0:number-1500, df.columns.get_loc(cols_to_normalize[0])] = 0
                df.iloc[0:number-1500, df.columns.get_loc(cols_to_normalize[1])] = 0

            # 对指定列进行归一化
            df[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])
            
            # 将 Sheet1 的内容写入 Excel writer 对象，并以当前 Excel 文件名命名工作表
            df.to_excel(writer, sheet_name=excel_file.split(',')[0], index=False)

    print(f"All sheets from {len(excel_files)} Excel files have been merged into {output_file}.")

input_folder = r"./data"
output_file = r"./data.xlsx"

# 合并 Excel 文件中的所有工作表
merge_excel_sheets(input_folder, output_file)
```

# 实际时序数据的贝叶斯自举

用于确定突变点time值，即确定cfc预测起始点

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as ss
from pyod.models.knn import KNN

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 12

def BayesianBootstrapcase(y,window_size, index1, index2):
    y = np.array(y)
    class BayesianBootstrap:
        def __init__(self, concentration: float = 1.):
            self.n_draws = 100000
            self.bins = 100
            self.concentration = concentration
        def sample(self, obs: np.ndarray, weights: np.ndarray = None):
            # If no weights passed, use uniform Dirichlet
            if weights is None:
                weights = np.ones(len(obs))
            # Normalize weights to mean concentration
            weights = weights / weights.mean() * self.concentration
            # Sample posteriors
            draws = ss.dirichlet(weights).rvs(self.n_draws)
            means = (draws * obs).sum(axis=1)
            vars = draws * (obs - means.reshape(self.n_draws, 1)) ** 2
            return means, np.sqrt(vars.sum(axis=1))
    data = y
    # 创建KNN模型
    clf = KNN()
    # 训练模型
    clf.fit(data.reshape(-1, 1))
    # 获取异常分数
    scores = clf.decision_scores_
    # print(scores)
    # 归一化分数
    normalized_scores = (scores - scores.min() + 1e-10) / (scores.max() - scores.min() + 1e-10)
    # 转化为权重
    weights = 1 - normalized_scores + 1e-10
    # print(weights)
    bbb = BayesianBootstrap()
    # 初始化用于存储每个窗口均值和标准差的列表
    means = []
    stds = []
    # 遍历每个时间窗口
    for i in range(window_size, len(data)):
        # 获取窗口内的数据和权重
        window_data = data[i - window_size:i]
        window_weights = weights[i - window_size:i]
        # 使用贝叶斯自举法计算窗口内的加权均值和标准差
        resampled_means, resampled_stds = bbb.sample(window_data, window_weights)
        # 记录这个窗口的均值和标准差
        means.append(resampled_means.mean())
        stds.append(resampled_stds.mean())
    def exponential_smoothing(data, alpha):
        smoothed_data = [data[0]]  # 初始平滑值为第一个数据点
        for i in range(1, len(data)):
            smoothed_value = alpha * data[i] + (1 - alpha) * smoothed_data[i - 1]
            smoothed_data.append(smoothed_value)
        return smoothed_data
    def normalize_list(lst):
        min_val = min(lst)
        max_val = max(lst)
        normalized = [(val - min_val) / (max_val - min_val) for val in lst]
        return normalized

    stds = stds[index1:index2]
    smoothed_data = exponential_smoothing(stds, alpha=0.3)
    smoothed_data = normalize_list(smoothed_data)
    smoothed_data.insert(0, 0)
    smoothed_data.append(0)
    return smoothed_data

# 提前准备好所有的word_number_pairs
word_number_pairs = {}
with open('peak_values.txt', 'r') as f:
    for line in f:
        word, number1, number2 = line.strip().split(' ')[0], line.strip().split(' ')[-2],line.strip().split(' ')[-1]
        word_number_pairs[word] = (int(number1),int(number2))

# xlsx_file = "./data.xlsx"
excel_data = pd.read_excel('data.xlsx', sheet_name=None)
count=0.3
with open('cfc.txt', 'w') as f:
    # print(excel_data.items)
    for sheet_name,df in excel_data.items():
        try:
            if sheet_name in word_number_pairs:
    #           data = pd.read_excel(xlsx_file, sheet_name=sheet_name)
                data = pd.read_excel('data.xlsx', sheet_name=sheet_name)
                
                y = data.iloc[:,2].tolist()
                window_size = 5
                index1, index2 = word_number_pairs[sheet_name] 
    #             print(f"index1:{index1},index2:{index2}")
                smoothed_data1 = BayesianBootstrapcase(y,window_size,index1-1500, index2-1500)
    #             print(data)
    #             print(f"smoothed_data1:{smoothed_data1}")
                fig = plt.figure(figsize=(9,4))
        
        
                ax = fig.add_subplot(121)
                #ax1.plot(data1['all'],color = '#D9A155')
                ax.set_xlim(0, data.shape[0])
                # plt.legend(handles=[l1], loc='upper left', labels=['Robert K. Merton'],fontsize=10
                #             ,frameon=False #去掉图例边框
                #             ,handlelength=0.8 #图例句柄的长度
                # )
                plt.fill_between([i for i in range(data.shape[0])],data['all'],-0.01,facecolor='#D9A155',alpha=0.2)
        
                ymin,ymax = ax.get_ylim()
        
                ax.set_ylim(ymin,ymax)
                ax.set_ylabel("Scientific fame",weight = 'bold')
                ax.set_xlabel("Year",weight = 'bold')
                ax.set_xticklabels([0,100,200,300,400,500], fontsize=10) # x轴字体设置
                ax.set_yticklabels([0, 0, 0.2, 0.4, 0.6, 0.8 ,1],fontsize=10) # y轴字体设置        
    #             print(len(smoothed_data1))
    #             print(index1+window_size)
    #             print(index2+window_size+2)
                plt.fill_between([i for i in range(index1+window_size,index2+window_size+2)],smoothed_data1,0,facecolor='#D9A155',alpha=0.45)
        
                max_value = max(smoothed_data1)
                max_index = smoothed_data1.index(max_value)
                count+=1
                print(f"{count}.{sheet_name}突变点为：{max_index+index1}")
    #             print(smoothed_data1.index(0))
                #plt.savefig('./Matthew effect_Matilda effect!_300dpi.jpg', bbox_inches='tight', dpi=300)
    #             plt.show()        
                ax.vlines(max_index+index1, ymin, ymax, 'black', '--',linewidth=0.8) # 垂直
                f.write(f"{sheet_name} {max_index+index1}\n")
        #         ax.vlines(506+2, ymin, ymax, 'black', '--',linewidth=0.8) # 垂直
        #         ax.vlines(329, ymin, ymax, 'gray', '--',linewidth=0.8) # 垂直
        except Exception as e:
            print(f"An error occurred in processing {sheet_name}: {e}")
            continue
print("finish!")
        # print(y2)
        #print(y2)
```

# cfc预测阶段

```python
import numpy as np
import tensorflow as tf

from tensorflow import keras
from ncps import wirings
from ncps.tf import CfC
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.font_manager import FontProperties
from matplotlib.patches import Rectangle
tf.random.set_seed(8)
plt.rcParams['font.family'] = 'Times New Roman'


def CfC_train(data_x, data_y, input_num, epochs_):
    # 创建一个由 8个全连接神经元组成的LTC/CfC网络，这些神经元接收1个输入特征的时间序列作为输入。此外，我们定义8个神经元中的1个作为输出
    fc_wiring = wirings.FullyConnected(8, 1)
    model = keras.models.Sequential(
        [
            keras.layers.InputLayer(input_shape=(None, input_num)),  # 输入的特征值个数
            CfC(fc_wiring, return_sequences=True),
        ]
    )
    model.compile(
        optimizer=keras.optimizers.Adam(0.01), loss=keras.losses.mean_squared_error, metrics = ['accuracy']
    )
    # model.summary()

    # 训练
    hist = model.fit(x=data_x, y=data_y, batch_size=8, epochs=epochs_, verbose=1,shuffle=False)
    # print(data_x)
    # print(data_y)

    # 训练loss值可视化
    # sns.set()
    # plt.figure(figsize=(6, 4))
    # plt.plot(hist.history["loss"], label="Training loss")
    # plt.legend(loc="upper right")
    # plt.xlabel("Training steps")
    # plt.show()

    return model


def model_prediction(time,time2,sheet_name,color):
    df = pd.read_excel("./data.xlsx", sheet_name=sheet_name)
    # 训练数据集
    # term
    data_y = np.array([df.iloc[row,1] for row in df.index][0:time]).reshape([1, time, 1]).astype(np.float32)
    # person
    data_x = np.stack(
        [
            np.array([df.iloc[row,2] for row in df.index][0:time]),
        ],axis=1
    )
    data_x = np.expand_dims(data_x, axis=0).astype(np.float32)
    # 测试数据集
    # term
    data_y_pre = np.array([df.iloc[row,1] for row in df.index]).reshape([1, df.shape[0], 1]).astype(np.float32)
    # person
    data_x_pre = np.stack(
        [
            np.array([df.iloc[row,2] for row in df.index]),
        ],axis=1
    )
    data_x_pre = np.expand_dims(data_x_pre, axis=0).astype(np.float32)
    # 模型训练
    epochs = 10
    model = CfC_train(data_y, data_x, 1, epochs)
    prediction = model(data_y_pre).numpy() # person
    prediction = np.where(prediction < 0, -prediction, prediction)

    draw(data_x_pre,prediction,sheet_name,time,time2,df,color)
    return data_x_pre[0, :, 0],prediction[0, :, 0]


def draw(real,prediction,sheet_name,time,time2,df,color):
    fig = plt.figure(figsize=(5, 3))
    # prediction = np.append(np.linspace(0,prediction[0, :, 0][245],245),prediction[0, :, 0][245:])
    ax = fig.add_subplot(111)
    ymin, ymax = ax.get_ylim()
    rect = Rectangle((time, ymin), time2 - time, ymax - ymin, facecolor='black', edgecolor='none', alpha=0.1)
    ax.add_patch(rect)
    ax.grid(False)
    # 真实数据
    # ax.plot(real[0, :, 0], label="GBNC", linewidth=1)
    plt.fill_between([i for i in range(len(real[0, :, 0]))], real[0, :, 0], -0.01, facecolor=color, alpha=0.13)
    # 预测数据
    # ax.plot(prediction[0, :, 0], label="CfC_output", linewidth=2)
    ax.plot(prediction[0, :, 0], linewidth=2, color = color)
    # ax.plot(input[0, :, 0], label="CfC_input")
    # ax.legend(loc="upper left"
    #           , fancybox=False
    #           , edgecolor='black'
    #           )
    ax.spines['top'].set_linewidth(0.9)
    ax.spines['left'].set_linewidth(0.9)
    ax.spines['right'].set_linewidth(0.9)
    ax.spines['bottom'].set_linewidth(0.9)
    ax.set_xlim(0, df.shape[0])
    if i==10 or i==11 or i==12:
        ax.set_xlabel('Year')
    if i==1 or i==4 or i==7 or i==10:
        ax.set_ylabel('Normalized Frenquency')
    ymin,ymax = ax.get_ylim()
    ax.vlines(time, ymin, ymax, 'black', '--',linewidth=0.6,alpha=0.2) # 垂直1
    ax.vlines(time2, ymin, ymax, 'black', '--', linewidth=0.6,alpha=0.2)  # 垂直2
    ax.set_ylim(ymin,ymax)
    title_obj = plt.title(sheet_name, weight = 'bold')
    title_obj.set_color('white')
    title_obj.set_bbox({'facecolor': color, 'alpha': 1, 'edgecolor': 'none'})
    font = FontProperties(size=14, weight='bold')
    title_obj.set_fontproperties(font)
    plt.savefig('./figures1/'+sheet_name+'_300dpi.jpg', bbox_inches='tight', dpi=300)
    plt.show()


# lists = ["Physicist","Microbe","Schizophrenia","Slow virus","Serendipity","Transistor","Henneman's size principle","Astrovirus","SARS","Stem cell","Vaccination","Matthew effect","Sheet2"]
# times = [140, 181, 211, 245, 245, 248, 264, 275, 303, 209, 181, 268,31] # 标记训练数据长度
# times2 = [222, 195, 312, 278, 314, 260, 282, 319, 306, 310, 202, 319,0] # 标记预测数据峰值
# #times2 = [222, 195, 312, 278, 314, 260, 282, 319, 306, 310, 0, 319] 
colors = ["#148758", "#963972", "#4EB0AE", "#1E6897","#876287","#E58D2E","#74AF4B","#4D78C8","#EDB117","#AD6239","#CE5745","#1E938D","#1E938D"]
lists = []
times = []
times2 = []

with open("./cfc.txt", "r") as file:
    for line in file:
        # 拆分每行，提取单词和数字
        line_parts = line.strip().split(" ")
        
        # 将第一个单词添加到单词列表中
        lists.append(line_parts[0])
        
        # 将最后的数字添加到数字列表中
        times.append(int(line_parts[-1])-1500)
        times2.append(int(line_parts[-1])-1500)




#i = 填写个数
for i in range(0,100):
    real_data = [] # 真实数据
    predict_data = [] # 预测数据
    list_data = [] # 需要预测的term
    time, time2, list, color = times[i], times2[i], lists[i], colors[1]
    real,predict = model_prediction(time, time2, list,color)
    real_data += real.tolist()
    predict_data += predict.tolist()
    list_data += [list for i in range(len(real.tolist()))][:]

    data=pd.DataFrame({'term':list_data,'real':real_data,'predict':predict_data})
    data.to_excel(f'./result./{lists[i]}.xlsx') # 保存预测结果
    print(lists[i])
    print("len")
    print(len(times), len(times2), len(lists))


#     print(real_data)
#     print(" ")
#     print([j-i for i,j in zip(times,times2)])


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as ss
from pyod.models.knn import KNN
import os
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 12

def BayesianBootstrapcase(y,window_size, index1, index2):
    y = np.array(y)
    class BayesianBootstrap:
        def __init__(self, concentration: float = 1.):
            self.n_draws = 100000
            self.bins = 100
            self.concentration = concentration
        def sample(self, obs: np.ndarray, weights: np.ndarray = None):
            # If no weights passed, use uniform Dirichlet
            if weights is None:
                weights = np.ones(len(obs))
            # Normalize weights to mean concentration
            weights = weights / weights.mean() * self.concentration
            # Sample posteriors
            draws = ss.dirichlet(weights).rvs(self.n_draws)
            means = (draws * obs).sum(axis=1)
            vars = draws * (obs - means.reshape(self.n_draws, 1)) ** 2
            return means, np.sqrt(vars.sum(axis=1))
    data = y
    # 创建KNN模型
    clf = KNN()
    # 训练模型
    clf.fit(data.reshape(-1, 1))
    # 获取异常分数
    scores = clf.decision_scores_
    # print(scores)
    # 归一化分数
    normalized_scores = (scores - scores.min() + 1e-10) / (scores.max() - scores.min() + 1e-10)
    # 转化为权重
    weights = 1 - normalized_scores + 1e-10
    # print(weights)
    bbb = BayesianBootstrap()
    # 初始化用于存储每个窗口均值和标准差的列表
    means = []
    stds = []
    # 遍历每个时间窗口
    for i in range(window_size, len(data)):
        # 获取窗口内的数据和权重
        window_data = data[i - window_size:i]
        window_weights = weights[i - window_size:i]
        # 使用贝叶斯自举法计算窗口内的加权均值和标准差
        resampled_means, resampled_stds = bbb.sample(window_data, window_weights)
        # 记录这个窗口的均值和标准差
        means.append(resampled_means.mean())
        stds.append(resampled_stds.mean())
#     print(len(stds))
    def exponential_smoothing(data, alpha):
        smoothed_data = [data[0]]  # 初始平滑值为第一个数据点
        for i in range(1, len(data)):
            smoothed_value = alpha * data[i] + (1 - alpha) * smoothed_data[i - 1]
            smoothed_data.append(smoothed_value)
        return smoothed_data
    def normalize_list(lst):
        min_val = min(lst)
        max_val = max(lst)
        normalized = [(val - min_val) / (max_val - min_val) for val in lst]
        return normalized
    
    stds = stds[index1:index2]
#     print(f"index1:{index1},index2:{index2}")
#     print(stds)
    smoothed_data = exponential_smoothing(stds, alpha=0.3)
    smoothed_data = normalize_list(smoothed_data)
    smoothed_data.insert(0, 0)
    smoothed_data.append(0)
    return smoothed_data

# 提前准备好所有的word_number_pairs
word_number_pairs = {}
with open('peak_values_predict.txt', 'r') as f:
    for line in f:
        word, number1, number2, number = line.strip().split('.')[0], line.strip().split(' ')[-2],line.strip().split(' ')[-1],line.strip().split(' ')[1]
        word_number_pairs[word] = (int(number1),int(number2),int(number))
print(word_number_pairs)
# 假设所有文件都在data文件夹中
data_folder = "./result"
file_names = os.listdir(data_folder)
count=0
with open('cfc_predict.txt', 'w') as f:
    # print(excel_data.items)
    for file_name in file_names:
#         print(file_name.split('.')[0])
        if file_name.split('.')[0] in word_number_pairs:
#             print("hello")
            index1, index2, number = word_number_pairs[file_name.split('.')[0]] 
#             data = pd.read_excel("./result/"+file_name.split('.')[0]+'.xlsx', skiprows=[i for i in range(1, number-1500)])
            data = pd.read_excel("./result/"+file_name.split('.')[0]+'.xlsx')
#             print(data)
            y = data.iloc[:,3].tolist()
#             print(f"len of y:{len(y)}")
            window_size = 5
            
#             print(f"index1:{index1},index2:{index2},number:{number}")
#             smoothed_data1 = BayesianBootstrapcase(y,window_size,index1-number, index2-number)
            smoothed_data1 = BayesianBootstrapcase(y,window_size,index1-1500, index2-1500)
#             print(data)
#             print(f"smoothed_data1:{smoothed_data1}")
#             fig = plt.figure(figsize=(9,4))
    
    
#             ax = fig.add_subplot(121)
            #ax1.plot(data1['all'],color = '#D9A155')
#             ax.set_xlim(0, data.shape[0])
            # plt.legend(handles=[l1], loc='upper left', labels=['Robert K. Merton'],fontsize=10
            #             ,frameon=False #去掉图例边框
            #             ,handlelength=0.8 #图例句柄的长度
            # )
#             plt.fill_between([i for i in range(data.shape[0])],data['predict'],-0.01,facecolor='#D9A155',alpha=0.2)
    
#             ymin,ymax = ax.get_ylim()
    
#             ax.set_ylim(ymin,ymax)
#             ax.set_ylabel("Scientific fame",weight = 'bold')
#             ax.set_xlabel("Year",weight = 'bold')
#             ax.set_xticklabels([0,100,200,300,400,500], fontsize=10) # x轴字体设置
#             ax.set_yticklabels([0, 0, 0.2, 0.4, 0.6, 0.8 ,1],fontsize=10) # y轴字体设置        
#             print(len(smoothed_data1))
#             print(index1+window_size)
#             print(index2+window_size+2)
#             plt.fill_between([i for i in range(index1+window_size,index2+window_size+2)],smoothed_data1,0,facecolor='#D9A155',alpha=0.45)
    
            max_value = max(smoothed_data1)
            max_index = smoothed_data1.index(max_value)
            count+=1
            print(f"{count}{file_name.split('.')[0]}突变点为：{max_index+index1}")
#             print(smoothed_data1.index(0))
            #plt.savefig('./Matthew effect_Matilda effect!_300dpi.jpg', bbox_inches='tight', dpi=300)
            plt.show()        
            ax.vlines(max_index+index1, ymin, ymax, 'black', '--',linewidth=0.8) # 垂直
           
            f.write(f"{file_name.split('.')[0]} {max_index+index1}\n")
    #         ax.vlines(506+2, ymin, ymax, 'black', '--',linewidth=0.8) # 垂直
    #         ax.vlines(329, ymin, ymax, 'gray', '--',linewidth=0.8) # 垂直
print("finish!")
        # print(y2)
        #print(y2)
```
